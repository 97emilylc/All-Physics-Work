# A simple SymPy example

First we import SymPy and initialize printing:

```python
from sympy import init_printing
from sympy import *
init_printing(use_unicode=True)
```

Create a few symbols:

```python
x,y,z,L,Lambda,a = symbols('x y z L, Lambda a')
```

Here is a basic expression:

```python
f = (((a-x)**2+y**2)**(-.5));f
```

```python
integrate(f,x)
```

```python
diff(e, x)
```

```python
integrate(e, z)
```

```python
from pylab import *
%matplotlib inline
t = arange(0.0, 2.0, 0.01)
s = sin(2*pi*t)
plot(t, s)

xlabel('time (s)')
ylabel('$ f(x) = {x}\omega [{m}{s}]$')
title('About as simple as it gets, folks')
grid(True)
show()
```
