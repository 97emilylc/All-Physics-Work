This notebook demonstrates some contour plot options for matplotlib.  Most notably, I wanted to play with chaning the thickness of the contour line depending on its value.

The contour documentation can be found here: http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contour

Make some sample data to play with
----------------------------------

```python
import numpy as np
from __future__ import division
from sympy import *
import matplotlib
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * 3.14, 100)
y = np.linspace(-3.14, 3.14, 100)
X, Y = np.meshgrid(x, y)
```

```python
D = np.sin(X ** 2) * np.cos(Y ** 2)
```

```python
plt.figure()
plt.contour(X, Y, D, cmap=cm.jet)
plt.show()
```

Change the linewith of the contours
-----------------------------------

```python
num_lines = len(line_widths)
print('Number of contours:  {0}'.format(num_lines))

line_widths = arange(20.) / 2
print('Line widths:  {0}'.format(line_widths))
```

```python
contour(X, Y, D, num_lines, linewidths=line_widths, cmap=cm.jet)
cb = colorbar()
cb.set_label('Python is great')
title('Changing width')
savefig('line-width.pdf')
```

```python
contourf(X, Y, D, len(line_widths), cmap=cm.jet)
title('...and with a filled contour')
```
